### Part A

In order to find the price $P$ of a bond for given yield $\lambda$, coupon payment $C$, face value $F$ and maturity date $t$ we have to apply the following formula:

$P = \frac{F}{(1+\lambda)^t} + \sum\limits_{k=1}^t \frac{C}{(1+\lambda)^k}$

```python
# calculates price given yield, coupon rate, face vale and maturity date
from __future__ import division
pv = lambda y,c,f,t: f/(1+y)**t + sum([(c*f)/(1+y)**k for k in range(1,t+1)])
```

```python
import matplotlib.pyplot as plt

y1 = [pv(yl/100,0.06,100,10) for yl in range(21)]
y2 = [pv(yl/100,0.10,100,10) for yl in range(21)]
y3 = [pv(yl/100,0.12,100,10) for yl in range(21)]
x = range(21)

figure()
plot(x,y1, 'r',label='Bond A')
plot(x,y2, 'g',label='Bond B')
plot(x,y3, 'b',label='Bond C')
pylab.legend(loc='upper right')

xlabel('Yield (%)')
ylabel('Price('+u'\u00A3'+')')
title('Price-Yield curve')
show()
```

#### Part B

We need to modify the formula in order to calculate the price for two payments per year:

$P = \frac{F}{(1+\frac{\lambda}{m})^{tm}} + \sum\limits_{k=1}^{tm} \frac{C/m}{(1+\frac{\lambda}{m})^k}$

```python
# calculates price given yield, coupon rate, face vale and maturity date
from __future__ import division
pvm = lambda y,m,c,f,t: f/(1+y/m)**(t*m) + sum([(c*f/m)/(1+y/m)**k for k in range(1,int(t*m)+1)])
```

```python
from ListTable import ListTable
table = ListTable()
table.append(['<b>Yield</b>', '<b>Price (bond A)</b>', '<b>Price (bond B)</b>','<b>Price (bond C)</b>'])
for y in xrange(21):
    bndA = "{0:.2f}".format(pvm(y/100,2,0.14,100,1/2))
    bndB = "{0:.2f}".format(pvm(y/100,2,0.14,100,1))
    bndC = "{0:.2f}".format(pvm(y/100,2,0.14,100,5/2))
    table.append([y,bndA,bndB,bndC])
table
```

```python
import matplotlib.pyplot as plt

y1 = [pvm(yl/100,2,0.14,100,1/2) for yl in range(21)]
y2 = [pvm(yl/100,2,0.14,100,1) for yl in range(21)]
y3 = [pvm(yl/100,2,0.14,100,5/2) for yl in range(21)]
x = range(21)

figure()
plot(x,y1, 'r',label='Bond A')
plot(x,y2, 'g',label='Bond B')
plot(x,y3, 'b',label='Bond C')
pylab.legend(loc='upper right')

xlabel('Yield (%)')
ylabel('Price('+u'\u00A3'+')')
title('Price-Yield curve')
show()
```
